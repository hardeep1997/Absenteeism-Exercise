## Age vs Probability

```python
#plt.subplots(figsize=(16,8))

sns.jointplot("Age", "Probability", data=test,
                  kind="reg", truncate=False,
                  xlim=(20,60), ylim=(0,10/10),
                  color="b", height=7)
```
